## List of dict ---> Dataframe pandas

```python
## import libraries needing
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## delete 'id'
df=[]
for i in range(len(data)):
    data1= data[i]
    del data1['_id']
    df.append(data1)
new_df=df
```

```python
len(new_df)
```

```python
df=pd.DataFrame(new_df)
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

#### I must to change the types of some features like : PassengerId, Survived, Pclass, Age, SibSp, Parch and Fare

```python
df[["PassengerId", "Survived"]] = df[["PassengerId","Survived"]].apply(pd.to_numeric)
df[["Pclass","Age","SibSp","Parch"]] = df[["Pclass","Age","SibSp","Parch"]].apply(pd.to_numeric)
df[["Fare"]] = df[["Fare"]].apply(pd.to_numeric)
df
```

```python
df.dtypes
```

```python
sns.pairplot(df)
```

```python
df.drop('Name',axis=1,inplace=True)
```

```python
df.drop('Ticket',axis=1,inplace=True)
```

```python
df
```

```python
df.corr()
```

```python
df=pd.get_dummies(df, columns=['Sex','Embarked'], drop_first=True)
df
```

```python
df.drop('PassengerId',axis=1,inplace=True)
```

```python
df
```

```python
X_Train=df[['Pclass','Age','SibSp','Parch','Fare','Sex_male','Embarked_Q','Embarked_S']]
Y_Train=df[['Survived']]
```

```python
X_Train,Y_Train
```

```python
Y_Train.T
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import learning_curve
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_Train,Y_Train)
```

```python
print(lr.score(X_Train,Y_Train))
```
